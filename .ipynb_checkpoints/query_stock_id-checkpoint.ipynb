### 上市

```python
from selenium import webdriver
from lxml import html
import time
import pandas as pd

# 初始化 Chrome WebDriver
driver = webdriver.Chrome()

# 設定目標網址
url = 'https://isin.twse.com.tw/isin/C_public.jsp?strMode=2'
driver.get(url)

# 獲取網頁的html
page_source = driver.page_source
tree = html.fromstring(page_source)

# 等待10秒，以確保網頁中的所有元素都已完全載入
time.sleep(10)

# 使用 XPath 抓取特定元素，這裡是表格中的第一列所有單元格
target_elements = tree.xpath('/html/body/table[2]/tbody/tr/td[1]')

# 檢查是否有找到元素
if target_elements:
    data = []  # 創建一個空列表來存儲解析的資料

    for element in target_elements:
        # 獲取元素的文字，如果為 None 則設為空字串
        text = element.text if element.text is not None else ""
        
        # 如果遇到特定的文字，則設置found為True並終止迴圈
        if text == "030001　富邦金群益42購01":
            found = True
            break
            
        # 檢查文字的前兩個字符是否為數字
        if text[:2].isdigit():
            # 使用空格分割文字，最多分割成兩部分
            parts = text.split(maxsplit=1)
            # 確保分割後有兩部分，然後將這些部分作為字典添加到列表中
            if len(parts) == 2:
                data.append({'id': parts[0], 'name': parts[1]})
    
    # 將列表轉換成 DataFrame 並列印出來
    df = pd.DataFrame(data)
    print(df)
        
else:
    # 如果沒有找到元素，輸出提示信息
    print("Elements not found")

# 關閉 WebDriver
driver.quit()
```

### 上櫃

```python
from selenium import webdriver
from lxml import html
import time
import pandas as pd

# 初始化 Chrome 瀏覽器驅動
driver = webdriver.Chrome()

# 設定欲訪問的網址
url = 'https://isin.twse.com.tw/isin/C_public.jsp?strMode=4'
driver.get(url)

# 獲取網頁源碼
page_source = driver.page_source

# 使用 lxml 解析 HTML
tree = html.fromstring(page_source)

# 等待一段時間，確保頁面中的所有元素都已完全加載
time.sleep(10)

# 使用 XPath 抓取特定的元素，此處抓取的是表格中第一列的所有單元格
target_elements = tree.xpath('/html/body/table[2]/tbody/tr/td[1]')

# 檢查是否成功找到目標元素
if target_elements:
    data = []  # 創建空列表以儲存數據
    start_collecting = False  # 設置一個標誌來控制數據收集的開始

    for element in target_elements:
        # 確保元素的文本不是 None，如果是 None 則設置為空字串
        text = element.text if element.text is not None else ""

        # 判斷是否達到開始收集數據的標記點
        if text == "1240　茂生農經":
            start_collecting = True

        # 如果達到結束收集數據的標記點，則退出迴圈
        if text == "8349A　恒耀甲特":
            break

        # 如果還未達到開始收集的點，則繼續下一次迴圈
        if not start_collecting:
            continue

        # 檢查元素文本的前兩個字符是否為數字
        if text[:2].isdigit():
            # 使用空格將文本分割為兩部分
            parts = text.split(maxsplit=1)
            # 確保分割後有兩部分，將這些部分以字典形式添加到數據列表中
            if len(parts) == 2:
                data.append({'id': parts[0], 'name': parts[1]})

    # 將收集到的數據轉換成 DataFrame 並列印
    df = pd.DataFrame(data)
    print(df)
else:
    # 如果沒有找到元素，輸出提示信息
    print("Elements not found")

# 關閉瀏覽器驅動
driver.quit()
```

```python
df.to_csv(r'D:\life_ease\stock_id\上櫃.csv')
```
